## Question 16
### Apply a function to the dataset to update the values of the "congested" to hold value 'Yes' if the value in the 'Num. of Veh. at that Hr.' column is greater than 10, else 'No' if otherwise.

```python
df = pd.read_csv('C:/Users/Ekemini Augustine/Downloads/test13.csv')
df
```

```python
def update_congestion(value):
    if value > 10:
        return 'Yes'
    else:
        return 'No'
```

```python
df['Congested'] = df['Num.of.veh.at.that.hr.'].apply(update_congestion)
df
```

```python
df = df.set_index('DateTime')
```

```python
df.to_csv('test16.csv')
```
